# WEB SCRAPPING

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
ws=pd.read_html(r"https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language")
```

```python
ws
```

```python
country1=ws[0] #for 1st table
```

```python
country1
```

```python
country2=ws[1]
```

```python
country2
```

```python
country1=country1.loc[:,["Alpha-3 code","Primary language?"]]
```

```python
country1
```

```python
country2=country2.loc[:,["Alpha-3 code","Primary language?"]]
```

```python
country2
```

Concatenating the countries

```python
master=pd.concat([country1,country2])
```

```python
master.head()
```

```python
master.reset_index(drop='index',inplace=True)
```

```python
master.head()
```

Renaming the Columns

```python
master.rename(columns={"Alpha-3 code":"country_code","Primary language?":"Official_lang"},inplace=True)
```

```python
master["Official_lang"]=master["Official_lang"].apply(lambda x:x.split()[0])
```

```python
master.head(2)
```

Converting to csv file

```python
master.to_csv('english_countries.csv',index=False)
```
